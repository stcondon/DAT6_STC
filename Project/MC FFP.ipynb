```python
def pointcalc(row, team):
    if row['winner'] == team:
        return 3
    elif row['winner'] == "Draw":
        return 1
    else:
        return 0
```

```python
def gdcalc(row, team = "Man City"):
    if row['winner'] == team:
        return abs(row['FTHG']-row['FTAG'])
    elif row['winner'] == "Draw":
        return 0
    else:
        return -abs(row['FTHG']-row['FTAG'])
```

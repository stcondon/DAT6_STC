```python
Todo = [Pre_cash, Post_cash]
Totes = pd.concat(Todo)
Totes.shape
```

```python
def pmaw(row):
    if row['FTR'] == 'H':
        return row['HomeTeam']
    elif row['FTR'] == 'A':
        return row['AwayTeam']
    else:
        return "Draw"
```

```python
def gdcalc(row, team = "Man City"):
    if row['winner'] == team:
        return abs(row['FTHG']-row['FTAG'])
    elif row['winner'] == "Draw":
        return 0
    else:
        return -abs(row['FTHG']-row['FTAG'])
```
